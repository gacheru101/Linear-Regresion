```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#created subplots to see the correlation between the TV, newspaper and radio vs sales
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))


axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

```python
#Assgning the feautures a VAR
#dropped the label y and were left with just the features X
X = df.drop('sales',axis=1) 
X.head()
```

```python
# assigning the label a VAR
y = df['sales']
```

```python
#importing train test spli model from sklearn
from sklearn.model_selection import train_test_split
```

```python
#It takes a certain % from the data based on the test and train
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)

#putting linear regression model we can import many models 
# we can look at help to look at our help to see the parameters 
from sklearn.linear_model import LinearRegression
```

```python
my_model = LinearRegression()
```

```python
my_model.fit(X_train,y_train)
```

```python
test_prediction = my_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df['sales'].mean()
```

```python
abs_mean =mean_absolute_error(y_test,test_prediction)
```

```python
sqr_mean = mean_squared_error(y_test,test_prediction)
```
